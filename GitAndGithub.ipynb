**What Is Git?**

Git is the distributed version control system used to track changes in files. Most commonly  on sourcecode. So, multiple can work on the same project safely and efficiently.

*Git help us...*

- Track changes to the files over time.

- Collaborate with other without overwritting work.

- Go back to the earlier version. if something break

- Create branches to work features independently.

**Repository..**

Repository is a place where all the version of our files and their complete change history are stored.

**working Directory(Local Repository or Git)  ----> Remote Repository on cloudserver(GitHub Is the remote)**

Working directory -----> Stage ------->Local Repository

To see the changes in the local repository we are going to stage the file to local repository.

 To track the any changes in the repository using **git status** command.

| File Type            | Description            |
| -------------------- | ---------------------- |
| Untracked            | Not added to Git (newly created file or folder)       |
| Tracked – Unmodified | No changes             |
| Tracked – Modified   | Changed but not staged |
| Tracked – Staged     | Ready to commit        |


To statge file or folder to the local repository using command..

- git add -A (To stage all the files and folder to working repository)

- git addd --all ( To stage all the file and folder to the local repository.)

To unstage all the changes to the initial stage by using the command .

- git reset
