**Constructor:**

A constructor in Python is a special method that is automatically called when an object of a class is created.
It is used to initialize (set up) the object’s data.
The constructor method is named __init__.
- __init__ is a constructor method in Python

- It runs automatically when a class object is created

- self refers to the current object

In Python, the keyword *pass* means “do nothing.”

It is used as a placeholder where a statement is required syntactically, but no action is needed yet.

Why pass is used:

- To avoid syntax errors when a block is required

- When you plan to write code later

- In empty functions, classes, loops, or conditionals
