**List slicing**

*Syntax*

list_name[start : end : step]

where start index is inclusive but end index is excluded. step is the interval between each number.
- *start* → starting index (inclusive)

- *end* → ending index (exclusive)

- *step* → gap between elements


```python
numbers = [0, 1, 2, 3, 4, 5]

numbers[1:4]     # [1, 2, 3]
numbers[:3]      # [0, 1, 2]
numbers[2:]      # [2, 3, 4, 5]
```

```python
#Negative slicing starts from the last elemnts of the list.
numbers[-4:-1]   # [2, 3, 4] 
```

**Step in Slicing**

```python
numbers[::2]     # [0, 2, 4]
numbers[1::2]    # [1, 3, 5]
```

**Reverse list**


```python
numbers[::-1] 
```

### Set

A set in Python is a collection of unordered, unique, and mutable elements.

- Unordered: Elements have no specific position; indexing is not allowed.

- Unique: Duplicate elements are automatically removed.

- Mutable: You can add or remove elements after creation.

- Data types allowed: Only immutable types like numbers, strings, and tuples can be elements of a set.



```python
# Using curly braces
my_set = {1, 2, 3}

# Using set() constructor
my_set2 = set([1, 2, 2, 3])
```

```python
my_set
```

```python
my_set2
```

**Key Features:**

- No duplicate values

- Supports mathematical operations like union, intersection, difference

- Useful for membership testing (in keyword)

- Cannot be indexed or sliced

```python
student_ids = [101, 102, 103, 101, 104]
unique_ids = set(student_ids)
print(unique_ids)  # Output: {101, 102, 103, 104}
```

**Find common items in the set(Intersection)**

```python
team_A = {"Alice", "Bob", "Charlie"}
team_B = {"Bob", "David", "Charlie"}

common_members = team_A & team_B
print(common_members)  # Output: {'Bob', 'Charlie'}
```

**Finds unique words in the documents**

```python
text = "python is easy and python is popular"
words = text.split()
unique_words = set(words)
print(unique_words)
```

**Union**

```python
day1 = {"Alice", "Bob", "Charlie"}
day2 = {"Bob", "Charlie", "David"}

all_attendees = day1 | day2  # Union
print(all_attendees)  # {'Alice', 'Bob', 'Charlie', 'David'}
```
